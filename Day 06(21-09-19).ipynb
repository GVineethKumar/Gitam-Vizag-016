## Visualizing Data
- An interseting way to display Dataframe
- with matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='blue')
plt.xlabel('x-Axis')
plt.ylabel('y-Axis')
plt.show()
```

```python
marks = [79,82,78,98]
avg_marks = [78,67,87,48]
divisions = ['A','B','C','D']
index = np.arange(len(divisions))
plt.bar(divisions,marks,color='green',width=0.7)
plt.show()
```

```python
plt.barh(divisions,marks,color='red')
plt.show()
```

```python
lang = ['python','java','c','c++']
rating = [75,70,55,65]
plt.pie(rating,labels = lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```

```python
import pandas as pd
filepath ='Match.csv'
def readCsvData(filepath):
    return pd.read_csv(filepath)
df = readCsvData(filepath)
print(df)
```

- 1.Display the count of matches based on the year wise

- 2.Plot a bar graph  no. of matches won by each team in the year  2011

- 3.Print  ManOfMatch count of each player in  Hyderabad stadium



```python
df["Season_Year"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
rows_2011 = df[df.Season_Year == 2011]
sns.countplot(y = rows_2011['match_winner'])
plt.xlabel('winner frequency')
plt.title('win data for the year 2011')
plt.show()
```

```python
Hyderabad=df[df.City_Name=='Hyderabad']
motm=pd.crosstab(index=Hyderabad["ManOfMach"],columns='count',margins=False)
print(motm)
```
